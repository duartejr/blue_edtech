# Pesquisa da discórdia

Import do pandas e leitura dos dados da pesquisa.

```python
import pandas as pd
df = pd.read_excel('dados.xlsx')
df.head()
```

Contabilizando a quantidade de valores nulos da planilha.

```python
print(df.info())
df.shape
```

Como visto em todas as colunas há valores nulos. Para filtrar parte desses valores será utilizado a coluna Nome. Pois é uma das com menores quantidade de falhas e maior quantidade de valores únicos.

```python
df = df.dropna(subset=['Nome'])
print(df.shape)
df.head()
```

A última coluna não têm informações úteis portanto será removida.

```python
df.drop('Unnamed: 11', axis=1, inplace=True)
print(df.shape)
df.head()
```

Primeiro veremos a distribuição espacial dos alunos.

```python
import matplotlib.pyplot as plt
#matplotlib inline

plt.figure(dpi=150)
df.groupby('Onde mora? (cidade/estado)').count()['Nome'].plot(kind='barh')
plt.title('Onde mora? (cidade/estado)')
```

## Distribuição espacial da turma por Estado
Algumas alterações são necessárias.

1º Substituir - por /

2º Adicionar estado a cidade de Rio de Janeiro

```python
# Trocando - por /
df['Onde mora? (cidade/estado)'].replace('-', '/', regex=True, inplace=True)
# Adicionando estado a cidade do Rio de Janeiro
mask = df['Onde mora? (cidade/estado)'].loc[df['Onde mora? (cidade/estado)'] == "Rio de Janeiro"]
df.loc[mask, 'Onde mora? (cidade/estado)'] = "Rio de Janeiro/RJ"
df['Onde mora? (cidade/estado)'].unique()
df['Estado onde mora'] = [x[1] for x in df['Onde mora? (cidade/estado)'].str.split('/')]
```

```python
#Extraindo estado
df['Estado onde mora'] = [x[1] for x in df['Onde mora? (cidade/estado)'].str.split('/')]
```

```python
plt.figure(dpi=150)
df.groupby('Estado onde mora').count()['Nome'].sort_values().plot(kind='barh')
plt.title('Estado onde mora')
```

## Histograma da data de nascimento

```python
plt.figure(dpi=150)
df['Data de Nascimento (DD/MM/AAAA)'].hist(figsize=(15,5))
plt.title('Data de Nascimento')
plt.xticks(rotation=45)
```

## Histograma time de futebol

```python
plt.figure(dpi=150)
df['Time de futebol?'].replace('não', 'Não', regex=True, inplace=True)
df['Time de futebol?'].replace('VAI CURINTIA', 'Corinthians', regex=True, inplace=True)
df.groupby('Time de futebol?').count()['Nome'].plot(kind='bar')
plt.xticks(rotation=90)
plt.title('Time de futebol?')
```

## Bolacha ou biscoito?

```python
df['Biscoito ou bolacha?'] = df['Biscoito ou bolacha?'].str.lower().str.rstrip()
plt.figure(dpi=150)
df['Biscoito ou bolacha?'].hist()
plt.xticks(rotation=45)
plt.title('Biscoito ou bolacha?')
```

## arroz por cima, por baixo ou do lado do feijão?

## Rodízio de carne, de pizza ou japa?
