## Introdução

O conjunto de dados de Câncer  de Mama está localizado na plataforma UCI no seguinte link - https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Coimbra. O objetivo deste conjunto de dados é classificar se um individuo possui ou não câncer de mama a partir destas variáveis obtidas através do exame de sangue.


Segue a descrição das variaveis:
- Age: idade em anos
- BMI: indice de massa corporal (kg/m2)
- Glucose: nivel de glicose no sangue (mg/dL)
- Insulin: nivel de insulina no sangue (µU/mL)
- HOMA: nivel de resistencia a insulina
- Leptin: leptina (ng/mL)
- Adiponectin: Adiponectina (µg/mL)
- Resistin: Resistina (ng/mL)
- MCP-1: nivel de proteina MCP-1 (pg/dL)
- Classification: é classificado como 1 sem câncer (grupo controle) e 2 com câncer (grupo paciente)

Para mais informações, segue o link do artigo: https://bmccancer.biomedcentral.com/articles/10.1186/s12885-017-3877-1

#### Referência: 
Patrício, M., Pereira, J., Crisóstomo, J., Matafome, P., Gomes, M., Seiça, R., & Caramelo, F. (2018). Using Resistin, glucose, age and BMI to predict the presence of breast cancer. BMC Cancer, 18(1).

Neste caso uma das primeiras etapaas a ser realizada é buscar-se entender como as variáveis fornecidas podem estar relacionadas com o caso que deseja-se resolver. A seguir são apresentadas breves descrições de cada uma das variáveis do Dataset extraidas diretamente dos links fornecidos.

**BMI** - O BMI em português IMC (ou índice de massa corporal) é um cálculo simples que permite avaliar se a pessoa está dentro do peso que é considerado ideal para a sua altura. O cálculo do IMC deve ser feito usando a seguinte fórmula matemática: Peso ÷ (altura x altura). Esta é uma fórmula simples que vários profissionais de saúde, incluindo médicos, enfermeiros e nutricionistas, usam para saber, de uma forma rápida, se a pessoa precisa ganhar ou perder peso. Isso porque, tanto o excesso de peso, como a falta, podem trazer diversas consequências para a saúde, desde desnutrição até ao aumento do risco de infartos ou AVC. Embora seja uma ferramenta muito comum, o IMC não é considerada a forma mais exata de avaliar o peso, já que não leva em consideração a composição corporal. Por isso, é comum que no caso de atletas (que possuem uma maior quantidade de massa muscular) seja aconselhado o uso de outras técnicas para uma avaliação mais detalhada do peso. Fonte: https://www.tuasaude.com/imc/

**Glucose** - A glicose é um tipo de açúcar, considerada a principal fonte de energia do organismo. Este açúcar é controlado por uma hormona produzida no pâncreas chamada insulina que ajuda a mover a glicose da corrente sanguínea. Se o pâncreas não produzir insulina suficiente ou se este não funcionar corretamente, a glicose pode acumular-se na corrente sanguínea, podendo provocar o surgimento de problemas em vários órgãos. Fonte: https://www.saudebemestar.pt/pt/exame/analises-clinicas/glicose/#:~:text=A%20glicose%20%C3%A9%20um%20tipo,para%20as%20c%C3%A9lulas%20do%20organismo.

**Insulin** - A insulina é produzida pelo pâncreas e funciona como uma “chave”, abrindo a células do corpo para que a glicose, proveniente do açúcar, entre e seja usada para gerar energia. Assim, a insulina também controla os níveis de glicose no sangue, que em excesso pode ser prejudicial, e está envolvida no armazenamento de gordura. A diabetes é caracterizada justamente pelos altos níveis de glicose no sangue. Pessoas diabéticas precisam tomar uma série de medidas para controlá-lo, como tomar injeções diárias de insulina artificial. Quando não controlada, a doença pode causar danos nos rins, nervos, visão, audição e no sistema cardiovascular. Ataques cardíacos, insuficiência renal e amputamento de membros são algumas das consequências mais graves e extremas. Fonte: https://www.danonenutricia.com.br/adultos/saude/insulina-definicao-tipos-como-aplicar

**HOMA** - O Índice de Homa é uma medida que aparece no resultado do exame de sangue que serve para avaliar a resistência à insulina (HOMA-IR) e atividade do pâncreas (HOMA-BETA) e, assim, auxiliar no diagnóstico da diabetes. A palavra Homa, significa Modelo de Avaliação da Homeostase e, geralmente, quando os resultados estão acima dos valores de referência, quer dizer que existe maior chance de desenvolver doenças cardiovasculares, síndrome metabólica ou diabetes tipo 2, por exemplo. Fonte: https://www.tuasaude.com/para-que-serve-o-indice-homa/

**Leptin** - É um dos hormônios reguladores do apetite mais conhecidos. Os experimentos iniciais com ratos demonstraram que a leptina é um hormônio anorexígeno e que um déficit de leptina poderia produzir um sinal orexígeno. Interessantemente, alguns experimentos mostraram um ciclo vicioso de resistência à leptina, no qual o hipotálamo se torna insensível à leptina. Estudos em modelos animais sugerem que a leptina exógena pode ser útil como terapia para perda de peso e antidiabetes. Alguns estudos mostraram que a leptina reduz os níveis de glicose em modelos de ratos com diabetes tipo 1 e tipo 2, indicando fortemente que o efeito é independente dos níveis de insulina. Vários trabalhos relataram também benefícios com o tratamento usando leptina em doenças hepáticas, amenorreia hipotalâmica, lipodistrofia e lipotoxicidade. Fonte:  https://www.labclasspardini.com.br/hormonios-associados-adiposidade-leptina-adiponectina-e-grelina/#:~:text=Adeponectina%20%C3%A9%20um%20horm%C3%B4nio%20associado,leptina%20e%20outras%20prote%C3%ADnas%20plasm%C3%A1ticas.

**Adiponectin** - Adeponectina é um hormônio associado com a obesidade e diabetes tipo 2. Como no caso da leptina, a adiponectina também é secretada pelo tecido adiposo. Porém, a adeponectina alcança níveis de concentração muito mais elevados do que a leptina e outras proteínas plasmáticas. Diferente da leptina, a concentração da adiponectina é inversamente associada com a quantidade de tecido adiposo, mesmo ela tendo um efeito anorexígeno similar ao da leptina. Concentrações de adiponectina podem não mudar em indivíduos obesos, a menos que uma perda de peso relativamente grande ocorra. Em modelos animais, a expressão excessiva de adiponectina previne o acúmulo lipídico hepático, fazendo com que os adipócitos proliferem. Semelhante aos efeitos da leptina, isto melhora a sensibilidade da insulina porque o lipídeo ectópico interfere com a sinalização da insulina e causa inflamação e apoptose. Estes processos, obviamente tem importantes implicações para as funções do fígado, células beta e dos rins. Ela também aumenta a beta oxidação dos ácidos graxos, diminui a lipogênese no fígado e músculo esquelético via fosforilação de ACC-1, e ajuda prevenir NAFLD de progredir para NASH em modelos animais. Outros aspectos observados em modelos animais foram: aumentos na angiogênese e na síntese de óxido nítrico. Notavelmente, a adiponectina é um potente sintetizador de insulina que aumenta a captação de glicose nos adipócitos, reduzindo a cascata de sinalização celular. Apesar da adiponectina estar associada à diversas doenças importantes, ainda não está claro como os clínicos deverão atuar diante de um resultado anormal, especialmente, um resultado baixo. Fonte: https://www.labclasspardini.com.br/hormonios-associados-adiposidade-leptina-adiponectina-e-grelina/#:~:text=Adeponectina%20%C3%A9%20um%20horm%C3%B4nio%20associado,leptina%20e%20outras%20prote%C3%ADnas%20plasm%C3%A1ticas.

**Resistin** - Resistina é uma proteína rica em cisteína secretada pelo tecido adiposo de mus e ratos. Em outros mamíferos, pelo menos primatas, porcos e cachorros, a resistina é secretada por células imunes e epiteliais. Tem como função o bloqueio da ação central da Leptina, hormônio produzido pelo tecido adiposo, que induz a saciedade. Este peptídeo esta relacionado a resistência à insulina e intolerância a glicose. Fonte: https://pt.wikipedia.org/wiki/Resistina

**MCP-1** - (Monocyte chemoattractant protein-1), também conhecido como Chemokine (CC-motivo) ligand 2 (CCL2), é da família de quimiocinas CC. Tem um papel vital no processo de inflamação, onde atrai ou melhora a expressão de outros fatores/células inflamatórias. Leva ao avanço de muitos transtornos por esse principal mecanismo de migração e infiltração de células inflamatórias como monócitos/macrófagos e outras citocinas no local da inflamação. O PCM-1 foi inculpatado na patogênese de inúmeras condições da doença, direta ou indiretamente, como o novo vírus corona, cânceres, doenças neuroinflamatórias, artrite reumatoide, doenças cardiovasculares. O elevado nível de PCM-1 tem sido observado em pacientes COVID-19 e comprovado ser um biomarcador associado à extremidade da doença juntamente com o IP-10. Fonte: https://www.sciencedirect.com/science/article/pii/S1567576921002344

```python
media_mcp1 = dataR2.groupby('Classification')['MCP.1'].mean()
diff_mcp1 = (media_mcp1[2] - media_mcp1[1]) / media_mcp1[2] * 100
print(f'Média MCP.1 pacientes sem câncer = {media_mcp1[1]:.2f}')
print(f'Média MCP.1 pacientes com câncer = {media_mcp1[2]:.2f}')
print(f'Diferenção percentual de MCP.1 entre os pacientes sem câncer = {diff_mcp1:.2f} %')
```

A distribuição dos pacientes segundo os índices de MCP.1 não ficou clara. Visualmente podemos inferir que as distribuições seriam similares e que quase não há diferença entre ambas mas seria interessante realizar um teste estatístico que validasse esta proposição.

Para verificar a se existe diferença significativa entre as distribuições foi escolhida estatística de soma de classificação de Wilcoxon para duas amostras. Disponível na biblioteca SciPy por meio do método `ranksums` contido no pacote `stats`.

Neste teste esatístico são feitas duas hipóteses a hipótese nula e a hipótese alternativa descritas as seguir:

**H<sup>nula</sup>**

A distribuição do valor de MCP-1 é a mesma para pacientes com e sem câncer.

**H<sup>alternativa</sup>**

A distribuição do valor de MCP-1 não é a mesma para pacientes com e sem câncer.

Portanto, este teste verifica se duas distribuições são iguais ou não.

Cálculo do teste de significância:

```python
#### VALE 0.6 pontos
from scipy.stats import ranksums

r = ranksums(pacientes_controle['MCP.1'], pacientes_cancer['MCP.1'])
print(f'O valor do p-value é {r.pvalue:.3f}.')
```

Como o resultado do p-value é maior que 0.05 não se pode descartar a hípotese nula. Portanto, o teste confirma que as distribuição de MCP-1 para pacientes com e sem câncer é a mesma. Pode-se então supor que esta variável não é relevante para identificar os pacientes com câncer de mama.

# Construção do modelo de Regressão Logística utilizando statsmodels para predizer a variável Classification

```python
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm

X = StandardScaler().fit_transform(dataR2.drop('Classification', axis=1))
y = dataR2['Classification'].map({1:0, 2:1})
X = sm.add_constant(X)
logit_model = sm.Logit(y, X)
result_logit = logit_model.fit()
```

Saída do seu modelo usando **seu_modelo.summary()**.

```python
result_logit.summary()
```

A partir do sumário estatístico do Statsmodel podemos selecionar as variáveis mais significativas para classificar os pacientes em positivos ou negativos. O valores da coluna `P>|z|` indica a probabilidade de a a variável x não ser significativa para prever y. Considerando-se isto podemos descartar as seguintes variáveis: **HOMA, Leptin, Adiponection**. Como na análise exploratória não foi identificada uma diferença significativa entre o nível de **MCP.1** entre os pacientes com e sem câncer esta variável também sera removida da modelagem. Na célula a seguir é realizado o mesmo teste estatístico com o Statsmodel e observa-se que o valor do Pseudo R², após a remoção das variáveis listadas, varia muito pouco em relação ao caso considerando todas as variáveis. 

```python
drops = ['Classification', 'HOMA', 'Leptin', 'Adiponectin', 'MCP.1']
X = StandardScaler().fit_transform(dataR2.drop(drops, axis=1))
y = dataR2['Classification'].map({1:0, 2:1})
X = sm.add_constant(X)
logit_model = sm.Logit(y, X)
result_logit = logit_model.fit()
result_logit.summary()
```

# Construção de modelo de Regressão Logística utilizando a biblioteca **sklearn**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

drops = ['Classification', 'HOMA', 'Leptin', 'Adiponectin', 'MCP.1']
X = StandardScaler().fit_transform(dataR2.drop(drops, axis=1))

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=.3, random_state=10)

logreg = LogisticRegression()
logreg.fit(X_treino, y_treino)
y_prev = logreg.predict(X_teste)

```

Matriz de confusão do modelo:

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

cm = confusion_matrix(y_teste, y_prev)
pd.DataFrame(cm, index = [['Observado', 'Observado'], ['Positivo', 'Negativo']],
                 columns = [['Predito', 'Predito'], ['Positivo', 'Negativo']])
```

Métricas a partir da matriz de confusão obtida (lembrando que a classe negativa é 1 - sem câncer e a classe positiva é 2 - com câncer)

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score


print('acurácia: {:.2f}'.format(accuracy_score(y_teste, y_prev)))
print('precisão: {:.2f}'.format(precision_score(y_teste, y_prev)))
print('especificidade: {:.2f}'.format(recall_score(y_teste, y_prev, pos_label=0)))
print('sensitividade: {:.2f}'.format(recall_score(y_teste, y_prev)))
```

Considerando-se que erros do tipo I são quando o modelo classifica verdadeiros positvos como negativos e erros do tipo II são quando o modelo classifica verdadeiros negativos como casos positivos. O modelo criado tender a cometer mais erros do tipo I. Apresenta valores de acurá e precisão relativamente elevados, porém, como trata-se de casos de doenças críticas a proposição feita serve apenas para fins didáticos, melhores modelos e formas de predição devem ser utilizados.
